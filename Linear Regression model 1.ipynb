```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
def targetfunc(x):
    return 3*x + 5
```

```python
x = np.linspace(-20, 50, 50)
fx = targetfunc(x)
plt.plot(x, fx)
plt.grid()
plt.show()
```

```python
np.random.seed(1)
t = fx + 50 * np.random.rand(len(x))
plt.plot(x, t, 'o')
plt.grid()
plt.show()
```

```python
def grad_loss(x, t, w0, w1):
    y = w0 * x + w1
    d_w0 = 2 * np.mean((y - t) + x)
    d_w1 = 2 * np.mean(y - t)
    return d_w0, d_w1

def grad_descent(x, t, w0, w1, lr, itr):
    _w0 = w0
    _w1 = w1
    eps = 0.1
    for i in range(1, itr):
        grad_w = grad_loss(x, t, _w0, _w1)
        _w0 = _w0 - lr*grad_w[0]
        _w1 = _w1 - lr*grad_w[1]
        if(max(np.absolute(grad_w))<eps):
            break
    return _w0, _w1

def show_line(x, t, w0, w1):
    # true - dot 
    plt.plot(x, t, 'o')
    #model - line
    y = w0 * x + w1
    plt.plot(x, y, color='red', linewidth = 4)
```

```python
w0 = 1.0
w1 = 1.0
lr = 0.001
itr = 1000
w0_opt, w1_opt = grad_descent(x, t, w0, w1, lr, itr)
print("W = {0:.3f}, {1:.3f}]".format(w0_opt, w1_opt))
show_line(x, t, w0_opt, w1_opt)
```
